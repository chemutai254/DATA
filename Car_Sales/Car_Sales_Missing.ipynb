```python
# Impute Colour with mode
df["Colour"] = df["Colour"].fillna(df["Colour"].mode()[0])
```

```python
# Impute Make with mode
df["Make"] = df["Make"].fillna(df["Make"].mode()[0])
```

```python
df.isna().sum()
```

```python
# df['Odometer (KM)'] = df['Odometer (KM)'].plot(kind='hist')
plt.figure(figsize=(8, 6))
sns.histplot(df['Odometer (KM)'], bins=20)
plt.title('Distribution of Odometer (KM)')
plt.xlabel('Odometer (KM)')
plt.ylabel('Frequency')
```

```python
# df['Doors'] = df['Doors'].plot(kind='hist')
plt.figure(figsize=(8, 6))
sns.histplot(df['Doors'], bins=10)
plt.title('Distribution of Doors')
plt.xlabel('Doors')
plt.ylabel('Frequency')
plt.show()
```

```python
# df['Price'] = df['Price'].plot(kind='hist')
plt.figure(figsize=(8, 6))
sns.histplot(df['Price'], bins=20)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Fill null values
df["Odometer (KM)"] = df["Odometer (KM)"].fillna(df["Odometer (KM)"].mean())
```

```python
# Fill null values
df["Doors"] = df["Doors"].fillna(df["Doors"].mean())
```

```python
# Fill null values
df["Price"] = df["Price"].fillna(df["Price"].median())
```
